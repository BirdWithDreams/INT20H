```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
orders_df = pd.read_csv('orders.csv')
nodes_df = pd.read_csv('nodes.csv')
```

```python
nodes_df
```

```python
orders_df.running_time = pd.to_datetime(orders_df.running_time)
orders_df.completed_time = pd.to_datetime(orders_df.completed_time)
orders_df
```

```python
orders_df['day_time'] = orders_df.running_time.dt.hour
orders_df
```

```python
orders_df['speed'] = orders_df.route_distance_km / orders_df.delta_time * 60
orders_df
```

```python
df = pd.merge(nodes_df, orders_df, on='Id')
df
```

```python
order_copy = orders_df.copy()
order_copy.index = order_copy['Id']
order_copy
```

```python
train_df = pd.concat([df.groupby('Id')['node_start'].count(),
           df.groupby('Id')['distance'].mean(), order_copy['route_distance_km'], order_copy['day_time'], order_copy['delta_time']], axis=1)
```

```python
train_df
```

```python
train_df = pd.concat([df.groupby('Id')['node_start'].count(), df.groupby('Id')['speed_x'].mean(),
           df.groupby('Id')['distance'].mean(), order_copy['route_distance_km'], order_copy['day_time'], order_copy['delta_time']], axis=1)
```

```python
train_df
```

```python
X = train_df.drop(columns=['delta_time']).to_numpy()
y = train_df['delta_time'].to_numpy()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape, X_val.shape, y_train.shape, y_val.shape # Only for training on a non-complete  
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
RFR = RandomForestRegressor()
param_grid = {'n_estimators': [64, 96, 128, 160, 192, 224, 256], 'max_depth': [3, 5, 7, 9, 11, 13]}
grid_search = GridSearchCV(estimator=RFR, param_grid=param_grid)
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_score = grid_search.best_score_

y_pred = grid_search.predict(X_val)

MSE = mean_squared_error(y_val, y_pred)
MAE = mean_absolute_error(y_val, y_pred)

print(f"Best: {best_params}, {best_score}. Result: MAE = {MAE}, RMSE = {np.sqrt(MSE)}")
```

```python
RFR = RandomForestRegressor(n_estimators=128, max_depth=7)

RFR.fit(X_train, y_train)

y_pred = RFR.predict(X_val)

MSE = mean_squared_error(y_val, y_pred)
MAE = mean_absolute_error(y_val, y_pred)

print(f"Best: {128, 9}, {best_score}. Result: MAE = {MAE}, RMSE = {np.sqrt(MSE)}")
```
